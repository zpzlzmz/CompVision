# 1.4.7 Canny Edge Detection

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('test.jpg',0)
edges = cv2.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python

img = cv2.imread('test.jpg')
img1 = cv2.pyrDown(img)
img2 = cv2.pyrDown(img)


cv2.imshow('image',img)
cv2.imshow('imag1e',img1)
cv2.imshow('image2',img2)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(img,100,200)
edges1 = cv2.Canny(img1,100,200)
edges2 = cv2.Canny(img2,100,200)
```

```python
cv2.imshow('image',edges)
cv2.imshow('imag1e',edges1)
cv2.imshow('image2',edges2)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
edge21 = cv2.pyrUp(edges)
edge22 = cv2.pyrUp(edge21)
```

```python
cv2.imshow('image',edges2)
cv2.imshow('image1',edge21)
cv2.imshow('image2',edge22)
cv2.waitkey(0)
cv2.destroyAllWindow()
```
