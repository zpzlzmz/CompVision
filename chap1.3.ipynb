```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
BLUE = [255,0,0]
img1 = cv2.imread('test.jpg')
```

```python
m=500
replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
```

```python
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1=cv2.imread('1.png')
img2=cv2.imread('2.jpg')
dst = cv2.addWeighted(img1,0.8,img2,0.3,0)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('dst')
cv2.createTrackbar('alpha','dst',0,255,nothing)
while(1):
    cv2.imshow('dst',dst)
    k=cv2.waitKey(1) &0xFF
    if k == 27:
        break
    alpha=cv2.getTrackbarPos('alpha','dst')
    dst = cv2.addWeighted(img1,alpha/255.,img2,1.-alpha/255.,0)
cv2.destroyAllWindows
```

```python
img1=cv2.imread('test.jpg')
e1 = cv2.getTickCount()
for i in range(5,49,2):
    img1 = cv2.medianBlur(img1,i)
e2=cv2.getTickCount()
t=(e2 - e1)/cv2.getTickFrequency()
print (t)
imageshow(img1)
```

```python
import cv2

def nothing(x):
    pass

cv2.namedWindow('Blur')
img1 = cv2.imread('tset.jpg')

cv2.createTrackbar('blur','Blur',5,49,nothing)

blur = 5

while(1):
    blur = cv2.getTrackbarPos('blur','Blur')
    if (blur % 2 == 0):
        blur = blur + 1
    img1 = cv2.medianBlur(img1,blur)
    
    cv2.imshow('Blur',img1)
    k=cv2.waitKey(1) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```
